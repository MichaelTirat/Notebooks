```python
offset = 0

annee_scolaire,academie,libelle_region_2016,code_region_2016,departement,code_departement,numero_de_l_ecole = [],[],[],[],[],[],[]
nom_etablissement,type_d_ecole,secteur_d_enseignement,nombre_d_eleves,adresse,code_postal,localite_acheminement = [],[],[],[],[],[],[]

for i in range(999):
    link = f'https://data.education.gouv.fr/api/explore/v2.0/catalog/datasets/fr-en-effectifs-premier-degre/records?limit=10&offset={offset}&include_app_metas=False&include_links=False'
    requete=requests.get(link)
    content = requete.json()


    for j in range(len(content['records'])):
        path = content['records'][j]['record']['fields']
        annee_scolaire.append(path['annee_scolaire'])
        academie.append(path['academie'])
        libelle_region_2016.append(path['libelle_region_2016'])
        code_region_2016.append(path['code_region_2016'])
        departement.append(path['departement'])
        code_departement.append(path['code_departement'])
        numero_de_l_ecole.append(path['numero_de_l_ecole'])
        nom_etablissement.append(path['nom_etablissement'])
        type_d_ecole.append(path['type_d_ecole'])
        secteur_d_enseignement.append(path['secteur_d_enseignement'])
        nombre_d_eleves.append(path['nombre_d_eleves'])
        adresse.append(path['adresse'])
        code_postal.append(path['code_postal'])
        localite_acheminement.append(path['localite_acheminement'])

    offset+=10
```

```python
dico_df = {"annee_scolaire":annee_scolaire,"academie":academie,"libelle_region_2016":libelle_region_2016,"code_region_2016":code_region_2016,
           "departement":departement,"code_departement":code_departement,"numero_de_l_ecole":numero_de_l_ecole,
           "nom_etablissement":nom_etablissement,"type_d_ecole":type_d_ecole,"secteur_d_enseignement":secteur_d_enseignement
           ,"nombre_d_eleves":nombre_d_eleves,"adresse":adresse,"code_postal":code_postal,"localite_acheminement":localite_acheminement}
```

```python
data = pd.DataFrame(dico_df)
```
